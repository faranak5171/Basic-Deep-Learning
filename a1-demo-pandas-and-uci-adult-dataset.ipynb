```python
import pandas as pd
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data.groupby(['sex']).size()
```

**2. What is the average age (*age* feature) of women?**

```python
data.loc[data['sex']=='Female', 'age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
(len(data.loc[data['native-country']=='Germany']) / len(data))* 100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
# You code here
print("The mean for more than 50K: ", data.loc[data['salary']=='>50K']['age'].mean())
print("The standard deviation for more than 50K: ", data.loc[data['salary']=='>50K']['age'].std())

print("\nThe mean for less than 50K: ", data.loc[data['salary']=='<=50K']['age'].mean())
print("The standard deviation for less than 50K: ", data.loc[data['salary']=='<=50K']['age'].std())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
# You code here
high_salary = len(data.loc[data['salary']=='>50K'])
high_salary_edu = len(data.loc[(data['salary']=='>50K') & (data['education'].isin(['Bachelors','Prof-school','Assoc-acdm','Assoc-voc','Masters','Doctorate']))])
print(high_salary==high_salary_edu)
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# You code here
df = pd.DataFrame(data.groupby(['race','sex']).describe())
print(df.head(5))
print(df[(df.index.get_level_values('race') == 'Amer-Indian-Eskimo') & (df.index.get_level_values('sex') == 'Male')]['age']['max'])
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
high_income = len(data.loc[data['salary']=='>50K'])
maried_men_high_income = len(data.loc[(data['salary']=='>50K') & (data['sex']=='Male') & (data['marital-status'].str.contains('Married'))])
single_men_high_income = high_income - maried_men_high_income
print("The portion of maried men who are earning a lot: ", maried_men_high_income)
print("The portion of single men who are earning a lot: ", single_men_high_income)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_hours = data['hours-per-week'].max()
num_max_hours = data.loc[data['hours-per-week'] ==max_hours].size
pct_high_income= data.loc[(data['hours-per-week'] ==max_hours) & (data['salary']=='>50K')].size
print("The max number of hourse a person works per week: {0}\nThe number of people works {0} hourse per week: {1}\nThe percentage of those who earn alot: {2:.2f}%"
      .format(max_hours, num_max_hours,(pct_high_income / num_max_hours)*100 ))
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
print('The average time of work (hours-per-week) for those who earn a lot ')
print(data.query('salary == ">50K"').groupby(['native-country'])['hours-per-week'].mean())
print('\n\nThe average time of work (hours-per-week) for those who earn a little ')
print(data.query('salary == "<=50K"').groupby(['native-country'])['hours-per-week'].mean())
```
